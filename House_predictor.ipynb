```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('AmesHousing.csv')
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
sns.histplot(df['SalePrice'],bins=10,kde=True)
```

```python
corr = df.select_dtypes(include=['int64','float64']).corr()

sale_corr = corr['SalePrice'].sort_values(ascending=False)

print(sale_corr.head(15))
```

```python
selected_columns = corr['SalePrice'][abs(corr['SalePrice']) >= 0.2].index.tolist()

print(selected_columns)
```

```python
df= df[['PID', 'Lot Frontage', 'Lot Area', 'Overall Qual', 'Year Built', 'Year Remod/Add', 'Mas Vnr Area', 'BsmtFin SF 1', 'Total Bsmt SF', '1st Flr SF', '2nd Flr SF', 'Gr Liv Area', 'Bsmt Full Bath', 'Full Bath', 'Half Bath', 'TotRms AbvGrd', 'Fireplaces', 'Garage Yr Blt', 'Garage Cars', 'Garage Area', 'Wood Deck SF', 'Open Porch SF','SalePrice']]
```

```python
df
```

```python
df.describe()
```

```python
X = df.drop(columns=['SalePrice'])
Y = df['SalePrice']
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
X_train, X_test, Y_train , Y_test= train_test_split( X,Y ,test_size=0.20, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_train_scaled
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
model = LinearRegression()
model.fit(X_train_scaled,Y_train)
```

```python
Y_pred = model.predict(X_test_scaled)
```

```python
Y_pred
```

```python
Y_test
```

```python
r2 = r2_score(Y_test,Y_pred)
r2
```

```python
n= X_test.shape[0]
p= X_test.shape[1]
adjusted_r2= 1-((1-r2)*(n-1))/(n-p-1)
print("Adjusted R2 score: ",adjusted_r2)
```

```python
mae = mean_absolute_error(Y_test, Y_pred)

mse = mean_squared_error(Y_test, Y_pred)
rmse= np.sqrt(mse)

print("MAE :", mae)
print("MSE:", mse)
print("RMSE:",rmse)
print("R2  :", r2)
```

```python
plt.scatter(Y_test, Y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from scipy.stats import randint
import numpy as np

# Parameter grid
param_dist = {
    'max_depth': randint(3, 30),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 10),
    'max_features': [None, 'sqrt', 'log2'],
    'criterion': ['squared_error', 'friedman_mse']
}

# Base model
model_dt_best = DecisionTreeRegressor()

# Randomized Search
random_search = RandomizedSearchCV(
    estimator=model_dt_best,
    param_distributions=param_dist,
    n_iter=50,
    cv=5,
    scoring='r2',
    random_state=42,
    n_jobs=-1
)

# Train
random_search.fit(X_train, Y_train)

# Best Model
dt_best = random_search.best_estimator_

# Predictions
Y_pred_dt_best = dt_best.predict(X_test)

# Evaluation
r2_dt_best = r2_score(Y_test, Y_pred_dt_best)
mae_dt_best = mean_absolute_error(Y_test, Y_pred_dt_best)
mse_dt_best = mean_squared_error(Y_test, Y_pred_dt_best)
rmse_dt_best = np.sqrt(mse_dt_best)

# Results
print("===== DT BEST (Randomized Search CV) =====")
print("Best Parameters:")
print(random_search.best_params_)

print("\nR2 Score :", r2_dt_best)
print("MAE      :", mae_dt_best)
print("MSE      :", mse_dt_best)
print("RMSE     :", rmse_dt_best)
```

```python
plt.scatter(Y_test, Y_pred_dt_best)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted")
plt.show()
```
